```python
matches = matches.drop(columns=['comp', 'notes'])
matches.head()
```

```python
matches['date'] = pd.to_datetime(matches['date'])
matches['target'] = (matches['result'] == 'W').astype('int')
matches.head()
```

```python
matches['venue_code'] = matches['venue'].astype('category').cat.codes
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
matches['hour'] = matches['time'].str.replace(':.+', '', regex=True).astype('int')
matches['day_code'] = matches['date'].dt.dayofweek
matches.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [10, 50, 100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10, 20, 30],
    'min_samples_leaf': [1, 2, 4, 8, 16]
}
```

```python
rf = RandomForestClassifier(random_state=1)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=1, verbose=2)
```

```python
train = matches[matches['date'] < '2023-01-01']
test = matches[matches['date'] > '2023-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
grouped_matches = matches.groupby('team')
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
matches_rolling = matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling.head()
```

```python
def make_predictions(data, predictors):
    train = data[data['date'] < '2023-01-01']
    test = data[data['date'] > '2023-01-01']
    grid_search.fit(train[predictors], train['target'])
    predictions = grid_search.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted = predictions), index = test.index)
    error = accuracy_score(test['target'], predictions)
    accuracy = precision_score(test['target'], predictions)
    return combined, error, accuracy
```

```python
combined, error, accuracy = make_predictions(matches_rolling, predictors + new_cols)
```
